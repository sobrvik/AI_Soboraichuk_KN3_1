```python
grades = np.random.randint(1, 13, 30)
print("All grades:", grades)
print("Average:", grades.mean())
print("Min:", grades.min())
print("Max:", grades.max())

final_grades = np.clip(grades + 2, 1, 12)
print("Adjusted grades:", final_grades)

failed_students = final_grades[final_grades < 6]
print("Failed students:", failed_students)
print("Count:", len(failed_students))
```
