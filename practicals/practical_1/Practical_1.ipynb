# Practical_1

Завдання: "Аналіз зарплат". Створіть масив зарплат 10 працівників (від 5000 до 20000).
Знайдіть середню зарплату.
Всі отримали премію 10%. (Підказка: помножити на 1.1).
Знайдіть, скільки людей тепер отримують більше 15000.


```python
import numpy as np

salaries = np.random.randint(5000, 20001, 10)
print("Зарплати працівників:", salaries)

average_salary = salaries.mean()
print("Середня зарплата:", average_salary)

bonus_salaries = salaries * 1.1
print("Зарплати після премії 10%:", bonus_salaries)

count_high_salary = np.sum(bonus_salaries > 15000)
print("Кількість працівників із зарплатою більше 15000:", count_high_salary)
```
